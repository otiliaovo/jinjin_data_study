```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
plt.rcParams['font.family'] = ['Source Han Sans CN']
```

```python
mask_data = pd.read_csv('mask_data.csv')
mask_data.info()
```

```python
mask_data = mask_data.dropna()
mask_data.info()
```

```python
mask_data = mask_data.drop_duplicates()
mask_data.describe()
```

```python
mask_data = mask_data[(mask_data["单价"] <= 200) & (mask_data["订单量"] > 0)]
mask_data.describe()
mask_data.info()
```

```python
mask_data
```

```python
# 转换日期数据，并设置对应的日期格式
date_data = pd.to_datetime(mask_data['日期'], format = '%Y-%m-%d')
# 提取月份数据
month_data = date_data.dt.month
#添加月份column
mask_data["月份"] = month_data
# 查看数据
mask_data
```

```python
# 写入csv
mask_data.to_csv("mask_data_clean.csv",index = False)
```

```python
province_month_sum = mask_data.groupby(['省',"月份"])['销售额'].sum()
province_month_sum.unstack()
```
